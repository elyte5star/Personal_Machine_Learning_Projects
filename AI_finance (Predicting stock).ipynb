# **Trading Rules Creation**

---
In this section, I developed two different trading rules, one is based on the RSI indicator (Relative Strength Index) and the second one is based on the MACD (Moving Average Convergence Divergence).

In order to obtain the best parameters for each rule, I decided to implement GA (Genetic Algorithm). This algorithm as mentioned above, returns the best combination of values for its fitness function (In our case the fitness function is the return using RSI or MACD).

There are different parameters we have to consider for each trading rules:

The parameters determined using GA for the RSI Based trading rule are:
* Number of periods for moving averages
* Upper Threshold
* Lower Threshold 

The parameters we determined using GA for the MACD Based trading rule are:
* Number of periods for fast moving average.
* Number of periods for slow moving average.
* Number of periods for signal moving average.

For the RSI trading rule, the GA implemented works using Real Values, because both thresholds allows it. However, for the MACD Based trading rule, the parameters must be naturals values/Integer (because it represents the number of periods for moving averages). As the GA does not have optimization of integers, we decided to work in binary, making the necessary conversions (as shown later in the report) to make the GA converge more quickly than using Real-Values. 




# **RSI BASED TRADING RULE**


Relative Strength Index (RSI): Measures the present changes in price to evaluate overbought or oversold conditions in the price of a stock.


**Key Setup of the RSI**


* Signals are considered overbought when the indicator is above 70%, and then one can sell.
* Signals are considered oversold when the indicator is above 30%, and then one can buy.
* The number of periods for the moving averages.

In the implementation of the trading rule, we have used a genetic algorithm to get the best values for these thresholds and the period for the moving averages.

https://www.investopedia.com/terms/r/rsi.asp

### **Load Train Data**

```python
getSymbols("MSFT",src="yahoo",from='2015-06-01',to='2019-08-01')
str(MSFT) 
gdatfr <- data.frame(date=index(MSFT), coredata(MSFT))

MAX_Price =max(max(gdatfr[,2]),max(gdatfr[,3]),max(gdatfr[,4]),max(gdatfr[,5]))
MIN_Price = min(min(gdatfr[,2]),min(gdatfr[,3]),min(gdatfr[,4]),min(gdatfr[,5]))
MAX_vol = max(gdatfr[,6])
MIN_vol = min(gdatfr[,6])
MAX_adj = max(gdatfr[,7])
MIN_adj = min(gdatfr[,7])

normalize <- function(x,i) {
  if (i=='open'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='high'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='low'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='close'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='vol'){
    return ((x - MIN_vol) / (MAX_vol -MIN_vol))
  }
  else if (i=='adj'){
    return ((x - MIN_adj) / (MAX_adj -MIN_adj))
  }
}
Open= normalize(Op(MSFT),'open')
High= normalize(Hi(MSFT),'high')
Low= normalize(Lo(MSFT),'low')
Close= normalize(Cl(MSFT),'close')
Volume= normalize(Vo(MSFT),'vol')
AdjClose= normalize(Ad(MSFT),'adj')
```

### **Define Fitness Function and perform GA on training Data**

```python
options(repr.plot.width=12, repr.plot.height=12)
set.seed(5)
trading_rule <- function(x) {
  stocks=0
  # DATA CONVERSION  (convert float from 0-1 to the range values): This is done in order to make GA converge faster
  # than setting the correct upper and lower bounds.Instead of 0-1
  days=as.integer(x[1]*20+10)  ##from 10 to 30
  upper_treshold=(x[2]*20+60)    ##from 60 to 80
  lower_treshold=(x[3]*20+20)       ##from 20 to 40

  ##RSI RULE IMPLEMENTATION
  rsi <- RSI(Close, days, maType="EMA")
  signal <- c() 
  signal [1:days+1] <- 0
  last_buy=0
  for (i in (days+1): length(Close)){
    if (rsi[i] < lower_treshold && stocks<1){ #buying threshold
      signal[i] <- -1
      stocks = stocks+ 1
      last_buy=i
    }else if(rsi[i] > upper_treshold && stocks>0){ #selling threshold
      signal[i] <- 1
      stocks =stocks - 1
    }else {  ##Do nothing
      signal[i] <- 0
    }
  }
  if (stocks >0){
      signal[last_buy] <- 0 ##then i dont count the last buy
      stocks =stocks - 1
  }
  signal<-reclass(signal,Close)
  trading <- Lag(signal)
  ret1 <- Cl(MSFT)*trading
  return (sum(ret1,na.rm=T))
}

library(GA)
GA <- ga(type = "real-valued",fitness = function(x) trading_rule(x),
         lower = c(0,0,0), upper = c(1,1,1), 
         maxiter = 2000, run = 50,pmutation = 0.2,popSize = 50)
summary(GA)
plot(GA)
```

# **MACD BASED TRADING RULE**

Moving Average Convergence/Divergence(MACD)-  is a trend and momentum indicator. It consists of two exponential moving averages(EMA) and a histogram, the MACD line itself is the difference between the 12-day(period) and 26-day(period) EMA, while the histogram is the difference between the MACD line and the Signal line). The Signal line(trigger line) is the 9-day EMA of the MACD line. These numbers are the default values. In the implementation, we have used a genetic algorithm to get the best values for our trading rule.Furthermore, we have implemented a strict buy/sell rule by combining both the supplementary rule and when the MACD crosses the Signal line.The supplementary rules give us the guarantee that the market is oversold before buying or overbought before selling.

**Key Setup of the MACD**


***Long trade setup***:


* MACD crosses below the Signal Line (supplementary rule)
* MACD crosses below the Zero Line (supplementary rule)
* Buy when the MACD crosses above its Signal line.

***Short trade setup***: 

* MACD crosses above the Signal Line (supplementary rule)
* MACD crosses above the Zero Line (supplementary rule)
* Sell when MACD  crosses below its Signal line.


https://www.tradingsetupsreview.com/macd-basic-trade-setup/





### **Load Train Data**

```python
getSymbols("MSFT",src="yahoo",from='2015-06-01',to='2019-08-01')
str(MSFT) 
gdatfr <- data.frame(date=index(MSFT), coredata(MSFT))
MAX_Price =max(max(gdatfr[,2]),max(gdatfr[,3]),max(gdatfr[,4]),max(gdatfr[,5]))
MIN_Price = min(min(gdatfr[,2]),min(gdatfr[,3]),min(gdatfr[,4]),min(gdatfr[,5]))
MAX_vol = max(gdatfr[,6])
MIN_vol = min(gdatfr[,6])
MAX_adj = max(gdatfr[,7])
MIN_adj = min(gdatfr[,7])

normalize <- function(x,i) {
  if (i=='open'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='high'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='low'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='close'){
    return ((x - MIN_Price) / (MAX_Price -MIN_Price))
  }
  else if (i=='vol'){
    return ((x - MIN_vol) / (MAX_vol -MIN_vol))
  }
  else if (i=='adj'){
    return ((x - MIN_adj) / (MAX_adj -MIN_adj))
  }
}
Open= normalize(Op(MSFT),'open')
High= normalize(Hi(MSFT),'high')
Low= normalize(Lo(MSFT),'low')
Close= normalize(Cl(MSFT),'close')
Volume= normalize(Vo(MSFT),'vol')
AdjClose= normalize(Ad(MSFT),'adj')
```

### **Define Fitness Function and perform GA on training Data**

```python
set.seed(10)#seed 10 works fine 5 works better

MACD_trading_rule_binary <- function(Small,Signal,High) {
  stocks=0
  Small=as.integer(Small)
  Signal=as.integer(Signal)
  High=as.integer(High)
  macd  <- MACD(Close, nFast=Small, nSlow=High, nSig=Signal, maType='EMA')#
  signal <- c() 
  signal [1:Signal+High+1] <- 0
  last_buy=0
  keep=0
  for (i in (Signal+High+1): length(Close)){
    keep=i
    if ( (macd$macd[i-1] < 0) && (macd$signal[i-1]<0) && ( macd$macd[i-1]<macd$signal[i-1]) &&  (macd$macd[i] > macd$signal[i]) && stocks<1 ){  #both lines below 0, and macd cross down-up signal line buy
      signal[i] <- -1
      stocks = stocks+ 1
      last_buy=i
    }else if((macd$macd[i-1] > 0 )&&  (macd$signal[i-1]>0) && ( macd$macd[i-1] > macd$signal[i-1] )&&  (macd$macd[i] < macd$signal[i]) && stocks>0){  #both lines above 0, and signal cross up-down macd line sell
      signal[i] <- 1
      stocks =stocks - 1
    }else {  ##Do nothing
      signal[i] <- 0
    }
  }
  if (stocks >0){
      signal[length(signal)] <- 1 ## At the end of the period sell the stock remaining
      stocks =stocks - 1
  }
  signal<-reclass(signal,Close)
  trading <- Lag(signal)
  ret1 <- Cl(MSFT)*trading
  return (sum(ret1,na.rm=T))
}

#Convert Binary to integer (THIS IS USED in order to make GA converge faster)
library(GA)
constraint <- function(x) {
  Signal <- c(x[1],x[2],x[3]) #Signal move from 6 to 13 
  Signal <- Reduce(function(x,y) x*2+y,Signal)
  Signal <- as.integer(Signal) + 5

  Small <- c(x[4],x[5],x[6]) ## fast move from 10 to 17 
  Small <- Reduce(function(x,y) x*2+y,Small)
  Small <- as.integer(Small) + 9

  High <- c(x[7],x[8],x[9],x[10]) ##slow limits 20 to 35 
  High <- Reduce(function(x,y) x*2+y,High)
  High <- as.integer(High) + 20

  sum <-MACD_trading_rule_binary(Small,Signal,High) 
  return(sum)
}
GAMACD <- ga(type = "binary", fitness = constraint, nBits = 10,pcrossover = 0.8,maxiter = 2000, run = 20,pmutation = 0.2,popSize = 50)
summary(GAMACD)
plot(GAMACD)
#end part2
```

# **Stock Price Prediction Using NeuralNetwork (neuralnet package)**

The NN implemented uses the market indicators (RSI,  MACD and  WMA) as inputs and also it use a lagged windows (5 days) of the previous Stock price to simulate a Recurrent Neural Network.

We focused on predicting the close price of the stock, as our trading rules generated before works only with the close price. Also it is important to mention that all the buy/sell signals obtained from the trading rules are Lagged(1) in order to buy on the next day, so it is not necessary to lag or do some modifications on the predicted value.

 The Neural Networks is configured with two hidden layers, the first with 10 neurons and the second one with 3 neurons

```python
means <-c()
stds <-c()
for (i in seq(1,length(DatasetMerged))){
  means<-c(means,mean(DatasetMerged[,i]))
  stds<-c(stds,sd(DatasetMerged[,i]))
}
DatasetMerged <- scale(DatasetMerged)
DatasetMerged <- as.data.frame(DatasetMerged)
```

###**Create Lag windows (in order to simulate RNN)**

```python
Create_Lag <-function(datas,Lag_list){
  myDf <-datas
  aux <- datas
  names<-names(DatasetMerged)
  aux_name <- names(DatasetMerged)
  for (i in Lag_list){
    lagged = aux %>% mutate_all(lag, n = i)
    myDf <- cbind(myDf, lagged$MSFT.Close)
    names<-c(names,paste(c('MSFT.Close',i),collapse=''))
  }
  names(myDf)<-names
  return (myDf)
}
lag_list = seq(1, 5)
DF = Create_Lag(DatasetMerged,lag_list)
DF <- DF[-(1:max(lag_list)), , drop = FALSE]
```

###**Remove Real data from the row we want to predict**

```python
CloseDF = DF[ , -which(names(DF) %in% c('MSFT.Low','MSFT.High'))]
dim(CloseDF)
```

###**Split Train and Test Data**

```python
train = CloseDF[1:1012,]  #1119 is the total (we have to remove the last 106 which are the test dataset)
test = CloseDF[1012:1118,]
n <- names(CloseDF)
```

###**Create Neural Network for Price forecasting**

```python
set.seed(6)
f <- as.formula(paste("MSFT.Close ~", paste(n[!n %in% "MSFT.Close"], collapse = " + ")))
```

### **Train Neural Network**

```python
nn <- neuralnet(f,data=train,hidden=c(10,3),linear.output=TRUE)#,act.fct = custom)
```

### **Performance withing Train Dataset**

```python
predicted.values <- neuralnet::compute(nn,train)
results <- data.frame(actual = (train$MSFT.Close*stds[4])+means[4] ,  prediction = (predicted.values$net.result*stds[4])+means[4])
print(paste0('RMSE Train Dataset: ', rmse(results$actual, results$prediction)))
```

### **Apply MACD Trading Rule**

```python
Apply_trading_rule <- function(Small,Signal,High,data) {
  stocks=0
  macd  <- MACD(data, nFast=Small, nSlow=High, nSig=Signal, maType='EMA')#
  signal <- c() 
  signal [1:Signal+High+1] <- 0
  last_buy=0
  keep=0
  for (i in (Signal+High+1): length(data)){
    keep=i
    if (stocks<1 && (macd$macd[i-1] < 0) && (macd$signal[i-1]<0) && ( macd$macd[i-1]<macd$signal[i-1]) &&  (macd$macd[i] > macd$signal[i])){  #both lines below 0, and macd cross down-up signal line buy
      signal[i] <- -1
      stocks = stocks+ 1
      last_buy=i
    }else if(stocks>0 && (macd$macd[i-1] > 0 )&&  (macd$signal[i-1]>0) && ( macd$macd[i-1] > macd$signal[i-1] )&&  (macd$macd[i] < macd$signal[i])){  #both lines above 0, and signal cross up-down macd line sell
      signal[i] <- 1
      stocks =stocks - 1
    }else {  ##Do nothing
      signal[i] <- 0
    }
  }
  if (stocks>0){
    print(paste0('Still have one stock which is current value is: ',Real_and_Predicted$Predicted[length(Real_and_Predicted$Predicted)]))
  }else{
    print('Dont have any stock left')
  }
  trading <- Lag(signal)
  ret1 <- data*trading
  return (sum(ret1,na.rm=T))
}
```

### **Final Return on Predicted Data using MACD**

### **Apply RSI Trading Rule**

```python

Apply_trading_rule2 <- function(days,lower_treshold,upper_treshold,data) {
  stocks=0
  rsi <- RSI(data, days, maType="EMA")
  signal <- c() 
  signal [1:days+1] <- 0
  last_buy=0
  for (i in (days+1): length(data)){
    if (rsi[i] < lower_treshold && stocks<1){  #buying treshold
      signal[i] <- -1
      stocks = stocks+ 1
      last_buy=i
    }else if(rsi[i] > upper_treshold && stocks>0){   #selling treshold
      signal[i] <- 1
      stocks =stocks - 1
    }else {  ##Do nothing
      signal[i] <- 0
    }
  }
  if (stocks>0){
    print(paste0('Still have one stock which is current value is: ',Real_and_Predicted$Predicted[length(Real_and_Predicted$Predicted)]))
  }else{
    print('Dont have any stock left')
  }
  signal<-reclass(signal,Close)
  trading <- Lag(signal)
  ret1 <- data*trading
  return (sum(ret1,na.rm=T))
}

```
